### 3. [2 puntos]

Usando expresiones regulares, escriba una función de Python que reciba una fecha en formato **Marzo 7, 2019** y retorne la fecha en formato **2019-07-03**

```python
def fecha(date):
    
    mes = {'Enero':"01", 'Febrero':"02", 'Marzo':"03", 'Abril':"04", 'Mayo':"05",\
           'Junio':"06", 'Julio':"07", 'Agosto':"08", 'Septiembre':"09", 'Octubre':"10",\
           'Noviembre':"11", 'Diciembre':"12"}
    
    Anio = re.findall('\d+$', date)
    
    Mes = re.findall('[a-zA-Z]+', date)
    
    Mes = mes[Mes[0]]
    
    Dia = re.findall('\d+(?=,)', date)    
    
    return "{Año}-{Dia}-{Mes}".format(Año=Anio[0], Dia=Dia[0], Mes=Mes)
```

```python
date = "Septiembre 21, 2019" 

# Imprimir fecha con formato Año-Dia-Mes:

fecha(date)
```

### 6. [10 puntos]


Vamos a hacer "scraping" a esta página: https://archive.ics.uci.edu/ml/datasets.php, que contiene un listado de 468 bases de datos que hacen parte del repositorio de la Universidad de California, Irvine.


Su tarea consiste en crear un "Pandas dataframe" que contenga 468 filas (una por base de datos) y las siguientes columnas:
- Nombre de la base de datos
- Link a la base de datos
- Tipo de datos
- Tipo de tarea a resolver (default task)
- Tipo de las variables
- Número de observaciones
- Número de variables
- Año
- Descripción de la base (Pista: Utilice la opción list view: https://archive.ics.uci.edu/ml/datasets.php?format=&task=&att=&area=&numAtt=&numIns=&type=&sort=nameUp&view=list)

Diviértase.

```python
from requests import get
from bs4 import BeautifulSoup as soup
import re
```

```python
url = 'https://archive.ics.uci.edu/ml/datasets.php'
html = get(url)
bases_datos = soup(html.content, 'html.parser')
```

```python
nombres_text = re.findall('(<b><a href="datasets/\w+.+?</a>)', str(bases_datos))
```

```python
import pandas as pd
# Crear dataframe a llenar
df = pd.DataFrame({'Columna inicializadora': [x for x in range(len(nombres_text))]})
```

### Info extraida de las bases de datos

```python
import pandas as pd

# Crear dataframe a llenar
df = pd.DataFrame({'Columna inicializadora': [x for x in range(len(nombres_text))]})

# Buscar info del dataframe (nombre/link)
df['nombres_text'] = re.findall('(<b><a href="datasets/\w+.+?</a>)', str(bases_datos))

# Filtrar la base para reducir el tiempo de procesamiento

df = df.loc[:50]

# Funcion para obtener el nombre

def get_name(x):
    nombre = re.search('(?<=>)(\w+.+)(?=<\/a>)', x).group()
    return nombre

# Funcion para obtener el link

def get_link(x):
    link = re.search('(?<="datasets/)(\w+.+)(?=">)',x).group()
    link = "https://archive.ics.uci.edu/ml/datasets/" + link
    return link

# Funcion para obtener tipo dato

def get_tipo_dato(x):
    tipo_dato = re.findall('(?<=<b>Data Set Characteristics:&nbsp;&nbsp;</b></p></td>\n\t\t<td><p class="normal">)\w+.+(?=</p></td>)', x)
    return tipo_dato

# Funcion para obtener el tipo de tarea a resolver

def get_tipo_tarea(x):
    tipo_tarea = re.findall('(?<=<b>Associated Tasks:</b></p></td>\n\t\t<td><p class="normal">)\w+.+(?=</p></td>)',str(x))
    return tipo_tarea

# Funcion para obtener el tipo variable

def get_tipo_var(x):
    tipo_var = re.findall('(?<=<p class="normal"><b>Attribute Characteristics:</b></p></td>\n\t\t<td><p class="normal">)\w+.+(?=</p></td>)',str(x))
    return tipo_var

# Funcion para obtener numero observaciones

def get_num_obs(x):
    num_obs = re.findall('(?<=<p class="normal"><b>Number of Instances:</b></p></td>\n\t\t<td><p class="normal">)\w+.+(?=</p></td>)',str(x))
    return num_obs

# Funcion para obtener numero variables

def get_num_var(x):
    num_var = re.findall('(?<=<b>Number of Attributes:</b></p></td>\n\t\t<td><p class="normal">)\w*.*(?=</p></td>)',str(x))
    return num_var

# Funcion para obtener el año

def get_año(x):
    año = re.findall('(?<=<b>Date Donated</b></p></td>\n\t\t<td><p class="normal">)\d*(?=-\d)',str(x))
    return año

# Funcion para obtener descripcion base

def get_bas_des(x):
    bas_des = re.findall('(?<=<b>Data Set Information:</b></p>\n<p class="normal">)\w+.+(?=</p>)',str(x))
    return bas_des


# Llenar columna del dataframe con info del nombre

df['nombre'] = df['nombres_text'].map(lambda x: get_name(x))

# Llenar columna del dataframe con info del link

df['link'] = df['nombres_text'].map(lambda x: get_link(x))

# Llenar columna del dataframe con texto de cada base de datos


df['texto_base'] = df['link'].map(lambda x: get(x).text)

# Llenar columna del dataframe con tipo dato

df['tipo_dato'] = df['texto_base'].map(lambda x: get_tipo_dato(x))

# Llenar columna del dataframe con tipo de tarea a resolver

df['tipo_tarea_resolver'] = df['texto_base'].map(lambda x: get_tipo_tarea(x))

# Llenar columna del dataframe con tipo variable

df['tipo_variable'] = df['texto_base'].map(lambda x: get_tipo_var(x))

# Llenar columna del dataframe con numero observaciones

df['numero_observaciones'] = df['texto_base'].map(lambda x: get_num_obs(x))

# Llenar columna del dataframe con numero variables

df['numero_variables'] = df['texto_base'].map(lambda x: get_num_var(x))

# Llenar columna del dataframe con el año

df['año'] = df['texto_base'].map(lambda x: get_año(x))

# Llenar columna del dataframe con descripcion base

df['descripcion_base'] = df['texto_base'].map(lambda x: get_bas_des(x))

# Eliminar columnas inecesarias


df
```
