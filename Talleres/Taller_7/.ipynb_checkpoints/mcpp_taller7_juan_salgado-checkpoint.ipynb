### 1.
Calcule el número de crímenes en cada Community Area en 2015. Haga un gráfico de barras que lo ilustre.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = [38.0, 20.0]


# ---> Leer base de datos:

base_2015 = pd.read_csv('Crimes_2015.csv', parse_dates=['Date'])
base_2015.head()
```

```python
# ---> Agrupar por 'Community Area':

base_com_area = base_2015.groupby("Community Area")
```

```python
# ---> Calcular total de ocurrencias en la base x 'Community Area':

com_area_count = base_com_area['ID'].agg('count').to_frame()
com_area_count.rename(columns = {'ID':'Ocurrencia crimenes'}, inplace = True)
    
    # Ordenando la base segun ocurrencias de crimen:
com_area_count = com_area_count.sort_values('Ocurrencia crimenes', ascending = False)
com_area_count.head()
```

```python
# ---> Graficar:

plt.rcParams["figure.figsize"] = [38.0, 20.0]

com_area_count.plot(kind='bar', color='green')
plt.xlabel('Community Area', fontsize = 30)
plt.legend(loc = 0, prop={'size': 30});
```

### 3.
Cree una tabla cuyas filas sean días del año (yyyy-mm-dd) y las columnas las 77 Community Areas. En cada campo de la tabla deberá haber el correspondiente número de crímenes. Seleccione algunas Community Areas que le llamen la atención y haga un gráfico de serie de tiempo.

Pista: El siguiente código puede serle útil.

```python
# Create function to strip time from date field, and use it to create another column
def to_day(timestamp):
    return timestamp.replace(minute=0,hour=0, second=0)

base_2015['Day'] = base_2015['Date'].apply(to_day)
```

```python
# ---> Agrupar por 'Community Area' y 'Date':

com_area_day = base_2015.groupby(['Community Area', 'Day'])

# ---> Totalizar por grupo:

com_area_count_day = com_area_day['ID'].agg('count').to_frame()
```

```python
# ---> Crear tabla con frecuencias diarias:

tabla_dias = pd.DataFrame({'Fecha': pd.date_range(start = '2015-01-01', end = '2015-12-31')})

# ---> Lista de 'Community Areas':

base_2015.rename(columns = {'Community Area':'Community_Area'}, inplace = True)

com_list = base_2015.Community_Area.unique()

# ---> Loop para llenar la tabla, para cada 'Community Area':

for com in com_list:
    
    for date in tabla_dias.Fecha:
        
        # try: si encuentra la fecha en la ocurrencia de crimenes
        try:
            
            asignar = com_area_count_day.loc[com].loc[date]['ID']
            tabla_dias.loc[tabla_dias.Fecha == date, com] = asignar
        
        # except: no encuentra la fecha en la ocurrencia de crimenes, asigna 'cero'
        except:
            
            asignar = 0
            tabla_dias.loc[tabla_dias.Fecha == date, com] = asignar
            
tabla_dias.head()
```

```python
# ---> Graficar las 'Community Areas' con mayor/menor ocurrencia total de crimenes:

# ---> Mayor ocurrencia (#25):

tabla_dias[25].plot()
```

```python
# ---> Menor ocurrencia (#9):

tabla_dias[9].plot()
```
