### 4.1   Exercises

1. Plot a simple graph of a sinus function in the range 0 to 3 with a step size of 0.01.

```python
# Generar variable entre el rango con el espacio requerido
variable = np.arange(0, 3, 0.01)
plt.plot(variable,np.sin(2*np.pi*variable))
```

2. Make the line red. Add diamond-shaped markers with size of 5.

```python
plt.plot(variable,np.sin(2*np.pi*variable), color = "red", marker = "D", markersize = 5)
```

3. Add a legend and a grid to the plot.

```python
plt.plot(variable,np.sin(2*np.pi*variable), color = "red", marker = "D", markersize = 5, label = 'Seno')
plt.legend()
plt.grid();
```
